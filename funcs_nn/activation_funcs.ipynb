#### ReLU

$f(x) = max(0,x)$
* Pro: tickle gradident vanishing when inference -> 1
* Cons: if parameter is 0, it will never be updated -> lots of neurons are dead

####  Swish activation

$f(x) = x*sigmoid(x)$

* Works better than ReLU on deeper model across a number of challenging tasks
* better for working with deep layers ~(40+ Layers). Adopted by VQGAN
